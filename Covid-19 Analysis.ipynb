#### Q1: TODO - Which country has the highest recorded Deaths

```python
highest_country_deaths = covid_country_wise_latest_df.sort_values('Deaths', ascending=False)
```

```python
highest_country_deaths.at[173, 'Country/Region']
```

The .at method used is useful to get a perticular value in a dataframe.

#### Q2: TODO - Which of the countries has the highest recovered cases.

```python
highest_recovered_df = covid_country_wise_latest_df.groupby('Country/Region')[['Recovered']].mean()
```

```python
highest_recovered_df.sort_values('Recovered', ascending=False)
```

We can see that the country with the highest recovered cases is Brazil followed by the US which is the country with the highest death.

#### Q3: TODO - Which countries had no deaths recorded

```python
lowest_death_perc_df = covid_country_wise_latest_df.groupby('Country/Region')[['Deaths / 100 Cases', 'Deaths']].mean()
```

```python
lowest_death_perc_df[lowest_death_perc_df['Deaths / 100 Cases'] == 0.00]
```

These Countries had no death recorded and, or no percentage of death.

#### Q4: TODO - Which country had the lowest confirmed cases recorded

```python
lowest_cases_df = covid_country_wise_latest_df.groupby('Country/Region')[['Confirmed']].mean()
```

```python
lowest_cases_df.sort_values('Confirmed', ascending=True)
```

The country with the lowest confirmed cases is "Western Sahara"
