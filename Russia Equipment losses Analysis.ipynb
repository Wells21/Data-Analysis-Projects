# Analysis of Russia Equipment Lossses

This Analysis would give us a clear visualization of how Russia Equipment are been destroyed during the war within 25th of Febraury to 24th of March.

lets import the neccessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (14, 10)
```

```python
russia_losses_equipment_df = pd.read_csv(r"C:\Users\Wells Praise\Desktop\Datasets\russia_losses_equipment.csv")
```

```python
russia_losses_equipment_df
```

# Data Preparation and Cleaning

```python
# dropping the special equipment column
russia_losses_equipment_df.drop(columns='special equipment', inplace=True)

russia_losses_equipment_df
```

```python
#Changing the column name (MRL)
russia_losses_equipment_new_df = russia_losses_equipment_df.rename(columns={'MRL': 'Multiple Rocket Launcher'})
```

```python
russia_losses_equipment_new_df
```

Lets expand the date column to year, month, days and weekdays

```python
russia_losses_equipment_new_df.date = pd.to_datetime(russia_losses_equipment_new_df.date)
```

```python
russia_losses_equipment_new_df['year'] = pd.DatetimeIndex(russia_losses_equipment_new_df.date).year
russia_losses_equipment_new_df['month'] = pd.DatetimeIndex(russia_losses_equipment_new_df.date).month
russia_losses_equipment_new_df['day'] = pd.DatetimeIndex(russia_losses_equipment_new_df.date).day
russia_losses_equipment_new_df['weekday'] = pd.DatetimeIndex(russia_losses_equipment_new_df.date).weekday
```

```python
russia_losses_equipment_new_df
```

We would be dropping the "Year" column, because we wouldnt be neding in for our analysis

```python
russia_losses_equipment_new_df.drop(columns = 'year', inplace=True)
```

```python
russia_losses_equipment_new_df
```

lets View some statistics of our dataset

```python
russia_losses_equipment_new_df.describe()
```

From this statistics it showed that Drones, Naval ships and Helicopters had the minimum loss in a particular day.
And also, Armored Personnel Carrier had the highest lost in a particular day.

# Exploratory Analysis

Lets find out the total number of aircraft, helicopter, tank, Armored Personnel Carrier, field artillery, Multiple Rocket Launcher, military auto, fuel tank, drone, naval ship and anti-aircraft warfare that was destroyed during the War between Febraury 25th to March 24th.

```python
total_lost_aircraft = np.array(russia_losses_equipment_new_df.aircraft).sum()

total_lost_helicopter = np.array(russia_losses_equipment_new_df.helicopter).sum()

total_lost_tank = np.array(russia_losses_equipment_new_df.tank).sum()

total_lost_apc = np.array(russia_losses_equipment_new_df['APC']).sum()

total_lost_field_artillery = np.array(russia_losses_equipment_new_df['field artillery']).sum()

total_lost_multiple_rocket_launcher = np.array(russia_losses_equipment_new_df['Multiple Rocket Launcher']).sum()

total_lost_military_auto = np.array(russia_losses_equipment_new_df['military auto']).sum()

total_lost_fuel_tank = np.array(russia_losses_equipment_new_df['fuel tank']).sum()

total_lost_drone = np.array(russia_losses_equipment_new_df['drone']).sum()

total_lost_naval_ship = np.array(russia_losses_equipment_new_df['naval ship']).sum()

total_lost_anti_aircraft = np.array(russia_losses_equipment_new_df['anti-aircraft warfare']).sum()
```

```python
print("total lost aircraft : {}".format(total_lost_aircraft))
print("total lost helicopter : {}".format(total_lost_helicopter))
print("total lost tank : {}".format(total_lost_tank))
print("total lost APC : {}".format(total_lost_apc))
print("total lost field artillery : {}".format(total_lost_field_artillery))
print("total lost multiple rocket launcher : {}".format(total_lost_multiple_rocket_launcher))
print("total lost military auto : {}".format(total_lost_military_auto))
print("total lost fuel tank : {}".format(total_lost_fuel_tank))
print("total lost drone : {}".format(total_lost_drone))
print("total lost naval ship : {}".format(total_lost_naval_ship))
print("total lost anti aircraft : {}".format(total_lost_anti_aircraft))
```

lets see the Trend.

```python
sns.set_style("darkgrid")
plt.title('Equipment Losses Trend')
plt.xlabel('Date')
plt.ylabel('Number of Equipment Lost')
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df.aircraft, 'o--g', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['helicopter'], 'x--y', mfc='b');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df.tank, 'v--b', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df.APC, '^--k', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['field artillery'], '*--c', mfc='b');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['Multiple Rocket Launcher'], 'd-.m', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['military auto'], 'D--g', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['fuel tank'], 'p--y', mfc='b');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['drone'], 'X--b', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['naval ship'], '|--k', mfc='r');
plt.plot(russia_losses_equipment_new_df.date, russia_losses_equipment_new_df['anti-aircraft warfare'], '4--c', mec='g');

plt.legend(['aircraft', 'helicopter', 'tank', 'APC', 'field artillery', 'MRL', 'millitary auto', 'fuel tank', 'drone', 'naval ship', 'anti-aircraft']);
```

This chart above shows us the Trend of the Russia equipment losses.
From the chart, we can see that the highest destroyed equipment is APC and the least is naval ship.
And also, the maximum number of lost equipment occured on the 24th of March 2022.

Lets Visualize the Equiment that had the highest lost.

```python
data = {'equipment' : ['aircraft', 'helicopter', 'tank', 'APC', 'field artillery', 'multiple rocket launcher', 'military auto', 'fuel tank', 'drone', 'naval ship', 'anti aircraft'],
                        'total_lost' : [1696, 2104, 9404, 32036, 4051, 1499, 16314, 1712, 315, 76, 766]
}
total_lost_equipment_df = pd.DataFrame(data)
```

```python
total_lost_equipment_df
```

```python
plt.title("Highest Equipment lost")
sns.barplot(total_lost_equipment_df.equipment, total_lost_equipment_df.total_lost);
plt.plot(total_lost_equipment_df.equipment, total_lost_equipment_df.total_lost, 'd--r', mfc='g', mec='g');
plt.xticks(rotation=60);
```

The chart above shows us the highest lost equipment which is APC.

Lets show the percentage of destroyed equipment within a month.

```python
plt.title('Percentage of equipment lost')
colors = sns.color_palette('pastel')[0:5]
plt.pie(total_lost_equipment_df.total_lost, labels = total_lost_equipment_df.equipment, colors = colors, autopct='%.0f%%', shadow=True);
plt.show();
```

APC has 46% lost equipment within the month.

We would be looking at which Weekdays has the most average lost equipment.

```python
grouped_weekday_df = russia_losses_equipment_new_df.groupby('weekday')[['aircraft', 'helicopter', 'tank', 'APC', 'field artillery', 'Multiple Rocket Launcher', 'military auto', 'fuel tank', 'drone', 'naval ship', 'anti-aircraft warfare']].mean()
```

```python
grouped_weekday_df
```

Each weekdays represents:
Monday = 0,
Tuesday = 1,
Wednessday = 2,
Thursday = 3,
Friday = 4,
Saturday = 5,
Sunday = 6.

Lets Visualize this with a Bar chart

```python
plt.figure(figsize = (20, 9))
weekdays = ['Monday', 'Tuesday', 'Wednessday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
width = 0.07
plt.title('average Lost equipment on Weekdays')
plt.xlabel('weekdays')
plt.ylabel('average lost equipment')

plt.bar(grouped_weekday_df.index, grouped_weekday_df.aircraft, width, label='aircraft', **{'hatch':'.', 'linewidth':3}, **{'ec':'g'}, **{'fc':'magenta'});
plt.bar(grouped_weekday_df.index+width, grouped_weekday_df.helicopter, width, label='helicopter', **{'hatch':'\\', 'linewidth':3}, **{'ec':'g'}, **{'fc':'black'});
plt.bar(grouped_weekday_df.index+width+width, grouped_weekday_df.tank, width, label='tank', **{'hatch':'*', 'linewidth':3}, **{'ec':'black'}, **{'fc':'red'});
plt.bar(grouped_weekday_df.index+width+width+width, grouped_weekday_df.APC, width, label='APC', **{'hatch':'/', 'linewidth':3}, **{'ec':'black'}, **{'fc':'blue'});
plt.bar(grouped_weekday_df.index+width+width+width+width, grouped_weekday_df['field artillery'], width, label='field artillery', **{'hatch':'+', 'linewidth':3}, **{'ec':'cyan'}, **{'fc':'green'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width, grouped_weekday_df['Multiple Rocket Launcher'], width, label='MRL', **{'hatch':'*', 'linewidth':3}, **{'ec':'black'}, **{'fc':'cyan'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width+width, grouped_weekday_df['military auto'], width, label='millitary auto', **{'hatch':'*', 'linewidth':3}, **{'ec':'black'}, **{'fc':'white'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width+width+width, grouped_weekday_df['fuel tank'], width, label='fuel tank', **{'hatch':'x', 'linewidth':3}, **{'ec':'b'}, **{'fc':'y'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width+width+width+width, grouped_weekday_df.drone, width, label='drone', **{'hatch':'o', 'linewidth':3}, **{'ec':'g'}, **{'fc':'b'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width+width+width+width+width, grouped_weekday_df['naval ship'], width, label='naval ship', **{'hatch':'O', 'linewidth':3}, **{'ec':'g'}, **{'fc':'magenta'});
plt.bar(grouped_weekday_df.index+width+width+width+width+width+width+width+width+width+width, grouped_weekday_df['anti-aircraft warfare'], width, label='anti-aircraft', **{'hatch':'.', 'linewidth':3}, **{'ec':'g'}, **{'fc':'orange'});

plt.legend()
plt.xticks(grouped_weekday_df.index, weekdays);
plt.show()
```

The given chart above shows that:
1. The highest average equipment loss occures on Thursdays. Which means there was more use of equipments on Thursdays.
2. On Fridays, equipments where less destroyed than any other weekdays.

# Conclusions and Inferences

From the Analysis, we drew the following conclusions:
1. APC equipment had the highest loss throughut Febraury 25th to March 24th. Which implies that APC was been used often for the War.
2. Naval ship had the lowest loss. This tells us that Naval ships were not really used often throughout that period.
3. Thursdays were the day the maximum loss occured in all the equipments, which implies that more equipments were used on Thursdays than any other weekdays.
