# Visualizing the Heights of Male and Female by Countries

Lets import the neccesary modules.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
matplotlib.rcParams['axes.edgecolor'] = 'black'
```

```python
heights_df = pd.read_csv(r'C:\Users\Wells Praise\Desktop\Datasets\Height of Male and Female by Country 2022.csv')
```

```python
from IPython.display import display
with pd.option_context('display.max_rows', 200):
    display(heights_df)
```

The Dataset is cleaned and processed for analysis already.

Lets look at the first 10 Countries with the highest average Male heights in cm

```python
# We group by country and we aggregrate using the mean() method
country_group_male_df = heights_df.groupby('Country Name')[['Male Height in Cm']].mean()
```

```python
country_group_male_df
```

```python
sorted_male_height_df = country_group_male_df.sort_values('Male Height in Cm', ascending=False).head(10)
```

```python
# the 10 countries with the highest male height in cm
sorted_male_height_df
```

lets check out the visualization using bar chart.

```python
plt.title('10 Countries with the highest Male heights');
sns.barplot(sorted_male_height_df['Male Height in Cm'], sorted_male_height_df.index);
```

from the graph above, the country with the highest male heights is Netherland.

Lets look at the first 10 Countries with the highest average female heights in cm

```python
country_group_female_df = heights_df.groupby('Country Name')[['Female Height in Cm']].mean()
```

```python
country_group_female_df
```

```python
sorted_female_height_df = country_group_female_df.sort_values('Female Height in Cm', ascending=False).head(10)
```

```python
sorted_female_height_df
```

lets show this on a bar chart.

```python
plt.title('10 countries with the highest Female Heights');
sns.barplot(sorted_female_height_df['Female Height in Cm'], sorted_female_height_df.index);
```

if we look at the similarities between the two charts, it tells us that Netherlands and Montenegro are the countries with the highest Male and Female heights. While,
Denmark is the third country that also has the highest Female heights.
Estonia is the third country that also has the highest male heights.

Lets make a line chart to show the trend of the Male and Female heights by Country

```python
plt.title('Relationship between the trend of the Male and Female heights')

plt.plot(sorted_male_height_df['Male Height in Cm'], 's-g');
plt.plot(sorted_female_height_df['Female Height in Cm'], 'o--r');
plt.legend(['Male', 'Female'])
plt.xticks(rotation=90);
```

The Trend above shows that the Highest Female hiegths is not above 172 cm

Lets show the trend of all the Male and Female heights

```python
sorted_all_male_height_df = country_group_male_df.sort_values('Male Height in Cm', ascending=False)
```

```python
sorted_all_male_height_df
```

```python
sorted_all_female_height_df = country_group_female_df.sort_values('Female Height in Cm', ascending=False)
```

```python
sorted_all_female_height_df
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 50));
plt.plot(sorted_all_male_height_df['Male Height in Cm'], sorted_all_male_height_df.index, marker='o', lw=2, ms=5, mfc='b', c='g');
plt.plot(sorted_all_female_height_df['Female Height in Cm'], sorted_all_female_height_df.index, 'o--r',  lw=2, ms=5, mfc='g');
plt.title('Trend of Male and Female Heights by country');
plt.legend(['Male', 'Female']);
```

# Conclusions and Findings
